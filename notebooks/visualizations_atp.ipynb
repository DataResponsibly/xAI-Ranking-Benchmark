```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
atp_time = pd.read_csv("results/time-experiment-ATP.csv", index_col=0)
atp_time.head()
```

```python
atp_time["fidelity_0"] = 1 - atp_time["fidelity_0"] / 86
atp_time["fidelity_1"] = 1 - atp_time["fidelity_1"] / 86
atp_time["fidelity_2"] = 1 - atp_time["fidelity_2"] / 86
```

```python
atp_time[atp_time["parameter"] == "sample_size"]
```

```python
atp_time.columns
```

```python
# average fidelity
atp_time["average_fidelity"] = atp_time[
    ["fidelity_0", "fidelity_1", "fidelity_2"]
].mean(axis=1)

# average agreement
atp_time["average_agreement_kendall"] = atp_time[
    ["agreement_kendall_0", "agreement_kendall_1", "agreement_kendall_2"]
].mean(axis=1)
atp_time["average_agreement_jaccard"] = atp_time[
    ["agreement_jaccard2_0", "agreement_jaccard2_1", "agreement_jaccard2_2"]
].mean(axis=1)
atp_time["average_agreement_euclidean"] = atp_time[
    ["agreement_euclidean_0", "agreement_euclidean_1", "agreement_euclidean_2"]
].mean(axis=1)
```

```python
atp_time[atp_time["parameter"] == "sample_size"]["parameter_value"].unique()
```

```python
atp_time[atp_time["parameter"] == "coalition_size"]["parameter_value"].unique()
```

```python
threads_data = atp_time[atp_time["parameter"] == "n_jobs"]

plt.figure(figsize=(7, 4))

approaches = threads_data["approach"].unique()

for approach in approaches:
    approach_data = threads_data[threads_data["approach"] == approach]
    plt.plot(
        approach_data["parameter_value"],
        approach_data["avg_time"],
        marker="o",
        label=approach,
    )

plt.xlabel("# threads")
plt.ylabel("Time")
plt.title("# threads vs Avg time")
plt.legend(title="Approach")
plt.show()
```

```python
threads_data = atp_time[atp_time["parameter"] == "n_jobs"]

plt.figure(figsize=(7, 4))

approaches = threads_data["approach"].unique()

for approach in approaches:
    approach_data = threads_data[threads_data["approach"] == approach]
    parameter_values = approach_data["parameter_value"]
    times = approach_data[["time_0", "time_1", "time_2"]]
    means = times.mean(axis=1)
    std_devs = times.std(axis=1)

    plt.errorbar(parameter_values, means, yerr=std_devs, fmt="o", label=approach)

plt.xlabel("# threads")
plt.ylabel("Time")
plt.title("# threads vs Time")
plt.legend(title="Approach")
plt.show()
```

```python
coalition_data = atp_time[atp_time["parameter"] == "coalition_size"]

plt.figure(figsize=(7, 4))

approaches = coalition_data["approach"].unique()

for approach in approaches:
    approach_data = coalition_data[coalition_data["approach"] == approach]
    plt.plot(
        approach_data["parameter_value"],
        approach_data["avg_time"],
        marker="o",
        label=approach,
    )

plt.xlabel("Coalition Size")
plt.ylabel("Time")
plt.title("Coalition Size vs Avg time")
plt.legend(title="Approach")

plt.xticks(coalition_data["parameter_value"].unique().astype(int))

plt.show()
```

```python
coalition_data = atp_time[atp_time["parameter"] == "coalition_size"]

plt.figure(figsize=(7, 4))

approaches = coalition_data["approach"].unique()

for approach in approaches:
    approach_data = coalition_data[coalition_data["approach"] == approach]
    parameter_values = approach_data["parameter_value"]
    times = approach_data[["time_0", "time_1", "time_2"]]
    means = times.mean(axis=1)
    std_devs = times.std(axis=1)

    plt.errorbar(
        parameter_values, means, yerr=std_devs, fmt="o", capsize=5, label=approach
    )

plt.xlabel("Coalition Size")
plt.ylabel("Time")
plt.title("Coalition Size vs Time")
plt.legend(title="Approach")

plt.xticks(coalition_data["parameter_value"].unique().astype(int))

plt.show()
```

```python
plt.figure(figsize=(7, 4))

for approach in approaches:
    approach_data = coalition_data[coalition_data["approach"] == approach]
    plt.plot(
        approach_data["parameter_value"],
        approach_data["average_fidelity"],
        marker="o",
        label=approach,
    )

plt.xlabel("Coalition Size")
plt.ylabel("Average Fidelity")
plt.title("Coalition Size vs Average Fidelity")
plt.legend(title="Approach")

plt.xticks(coalition_data["parameter_value"].unique().astype(int))

plt.show()
```

```python
plt.figure(figsize=(7, 4))

coalition_rank_data = atp_time[
    (atp_time["parameter"] == "coalition_size") & (atp_time["approach"] == "rank")
]

means = []
std_devs = []
coalition_sizes = sorted(coalition_rank_data["parameter_value"].unique())

# mean and standard deviation for each coalition size
for coalition_size in coalition_sizes:
    fidelity_values = coalition_rank_data[
        coalition_rank_data["parameter_value"] == coalition_size
    ][["fidelity_0", "fidelity_1", "fidelity_2"]]
    values = fidelity_values.values.flatten()
    means.append(np.mean(values))
    std_devs.append(np.std(values))

plt.figure(figsize=(7, 4))

plt.errorbar(coalition_sizes, means, yerr=std_devs, fmt="o", label="Fidelity")

plt.xlabel("Coalition Size")
plt.ylabel("Fidelity")
plt.title("Coalition Size vs Fidelity (rank approach)")
plt.xticks(ticks=coalition_sizes, labels=coalition_sizes)

plt.show()
```

```python
plt.figure(figsize=(7, 4))

coalition_rank_data = coalition_data[coalition_data["approach"] == "rank"]
for agreement in ["kendall", "jaccard", "euclidean"]:
    plt.plot(
        coalition_rank_data["parameter_value"],
        coalition_rank_data[f"average_agreement_{agreement}"],
        marker="o",
        label=agreement,
    )

plt.plot(
    coalition_rank_data["parameter_value"],
    coalition_rank_data["average_fidelity"],
    marker="o",
    label="fidelity",
)

plt.xlabel("Coalition Size")
plt.ylabel("Average Agreement")
plt.title("Coalition Size vs Average Agreement (rank approach)")
plt.legend(title="Method")

plt.xticks(coalition_data["parameter_value"].unique().astype(int))

plt.show()
```

```python
plt.figure(figsize=(7, 4))

for agreement in ["kendall", "jaccard2", "euclidean"]:
    means = coalition_rank_data[
        [
            f"agreement_{agreement}_0",
            f"agreement_{agreement}_1",
            f"agreement_{agreement}_2",
        ]
    ].mean(axis=1)
    std_devs = coalition_rank_data[
        [
            f"agreement_{agreement}_0",
            f"agreement_{agreement}_1",
            f"agreement_{agreement}_2",
        ]
    ].std(axis=1)
    plt.errorbar(
        coalition_rank_data["parameter_value"],
        means,
        yerr=std_devs,
        fmt="o",
        label=f"agreement_{agreement}",
    )

means = coalition_rank_data[["fidelity_0", "fidelity_1", "fidelity_2"]].mean(axis=1)
std_devs = coalition_rank_data[["fidelity_0", "fidelity_1", "fidelity_2"]].std(axis=1)
plt.errorbar(
    coalition_rank_data["parameter_value"],
    means,
    yerr=std_devs,
    fmt="o",
    label="fidelity",
)

plt.xlabel("Coalition Size")
plt.ylabel("Agreement/Fidelity")
plt.title("Coalition Size vs Agreement/Fidelity (rank approach)")
plt.legend(title="Method")

plt.xticks(coalition_rank_data["parameter_value"].unique().astype(int))

plt.show()
```
